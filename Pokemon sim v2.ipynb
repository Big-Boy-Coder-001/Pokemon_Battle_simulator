```python
def randomgen():
    if not p1moves:
        global poke_moveset
        poke_moveset=p1['Moves'][0]
        poke_moveset=ast.literal_eval(poke_moveset)
        poke_moveset=pd.DataFrame.from_dict(poke_moveset)
        poke_moveset=poke_moveset.T
        poke_moveset=poke_moveset.reset_index()
        poke_moveset['Power']=poke_moveset['Power'].replace('--',0)
        poke_moveset['Power']=poke_moveset['Power'].replace('??',40)
        poke_moveset['Accuracy']=poke_moveset['Accuracy'].replace('--',100)
        poke_moveset['Accuracy']=poke_moveset['Accuracy'].replace('??',0)

        if (len(poke_moveset)>3):
            n=random.sample(range(0, len(poke_moveset)), 4)
            a=(poke_moveset.loc[poke_moveset.index==n[0]])
            b=(poke_moveset.loc[poke_moveset.index==n[1]])
            c=(poke_moveset.loc[poke_moveset.index==n[2]])
            d=(poke_moveset.loc[poke_moveset.index==n[3]])
            p1moves.append(a)
            p1moves.append(b)
            p1moves.append(c)
            p1moves.append(d)
        else:
            x=len(poke_moveset)
            n=random.sample(range(0,len(poke_moveset)),x)
            for i in range(x):
                temp=(poke_moveset.loc[poke_moveset.index==n[i]])
                p1moves.append(temp)
    if not p2moves:
        poke_moveset2=p2['Moves'][0]
        poke_moveset2=ast.literal_eval(poke_moveset2)
        poke_moveset2=pd.DataFrame.from_dict(poke_moveset2)
        poke_moveset2=poke_moveset2.T
        poke_moveset2=poke_moveset2.reset_index()
        poke_moveset2['Power']=poke_moveset2['Power'].replace('--',0)
        poke_moveset2['Power']=poke_moveset2['Power'].replace('??',40)
        poke_moveset2['Accuracy']=poke_moveset2['Accuracy'].replace('--',100)
        poke_moveset2['Accuracy']=poke_moveset2['Accuracy'].replace('??',0)
        if (len(poke_moveset2)>3):
            n=random.sample(range(0, len(poke_moveset2)), 4)
            a=(poke_moveset2.loc[poke_moveset2.index==n[0]])
            b=(poke_moveset2.loc[poke_moveset2.index==n[1]])
            c=(poke_moveset2.loc[poke_moveset2.index==n[2]])
            d=(poke_moveset2.loc[poke_moveset2.index==n[3]])
            p2moves.append(a)
            p2moves.append(b)
            p2moves.append(c)
            p2moves.append(d)
        else:
            x=len(poke_moveset2)
            n=random.sample(range(0,len(poke_moveset2)),x)
            for i in range(x):
                temp=(poke_moveset2.loc[poke_moveset2.index==n[i]])
                p2moves.append(temp)
```
